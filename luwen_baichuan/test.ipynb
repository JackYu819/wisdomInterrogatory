```python
import sys
from transformers import AutoModelForCausalLM, AutoTokenizer
import os
```

```python
os.environ["CUDA_VISIBLE_DEVICES"] = "4"
tokenizer = AutoTokenizer.from_pretrained("output/baichuan_80k/", trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained("output/baichuan_80k/", device_map="auto", trust_remote_code=True)
```

```python
model.generation_config
```

```python
prompt = "在上班路上脚踝关节骨折算不算是工伤，企业应该负什么责任？"
inputs = tokenizer(f'</s>Human:{prompt} </s>Assistant: ', return_tensors='pt')
inputs = inputs.to('cuda')
pred = model.generate(**inputs, max_new_tokens=300, repetition_penalty=1)
print(tokenizer.decode(pred.cpu()[0], skip_special_tokens=True))
```
